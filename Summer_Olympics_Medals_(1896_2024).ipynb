# Thế vận hội Olympic mùa hè từ Thế vận hội Athens khai mạc năm 1896 đến Thế vận hội Paris 2024

Features:
*   player_id: Mã định danh duy nhất cho mỗi vận động viên.
*   Name: Tên của vận động viên.
*   Sex: Giới tính của vận động viên (Nam (M) /Nữ (F)).
*   Team: Quốc gia mà vận động viên đại diện.
*   NOC: Mã Ủy ban Olympic Quốc gia của quốc gia.
*   Year: Năm diễn ra Thế vận hội Olympic.
*   Season: Mùa của Thế vận hội Olympic (Mùa hè).
*   City: Thành phố đăng cai Thế vận hội Olympic.
*   Sport: Thể loại thể thao mà vận động viên thi đấu.
*   Event: Sự kiện cụ thể trong môn thể thao.
*   Medal: Loại huy chương được trao (Gold, Silver, Bronze) hoặc “Không có huy chương” nếu không được trao.

Việc tăng số lượng các môn thể thao thi đấu cũng đồng nghĩa việc tặng số lượng huy chương qua mỗi mùa Olympic

Tổng số huy chương qua các năm: Số lượng huy chương thường tăng theo thời gian, đặc biệt là từ giữa thế kỷ 20 khi Olympic trở nên toàn cầu hơn và có sự gia tăng số lượng sự kiện và quốc gia tham gia.

```python
olympics_df_medal = data[data['Medal'] != 'No medal']
olympics_df_medal
```

```python
olympics_df_medal.Medal.value_counts()
```

```python
medals_count = olympics_df_medal.groupby(['Team', 'Medal']).size().unstack(fill_value=0)


top_10_countries = medals_count.sum(axis=1).nlargest(10).index
top_10_medals = medals_count.loc[top_10_countries]

colors = ['#FFD700', '#C0C0C0', '#CD7F32']
top_10_medals.plot(kind='bar', stacked=True, figsize=(10, 7), color=colors)
plt.title('Top 10 Quốc Gia Đạt Huy Chương Nhiều Nhất')
plt.xlabel('Quốc Gia (Team)')
plt.ylabel('Số Huy Chương')
plt.legend(title='Loại Huy Chương')
plt.show()
```

Sử dụng biểu đồ thể hiện các quốc gia có số lượng huy chương cao nhất. Biểu đồ này giúp so sánh thành tích của các quốc gia hàng đầu qua nhiều kỳ Olympic.

Hoa Kỳ: Là quốc gia dẫn đầu về tổng số huy chương, Hoa Kỳ đã thống trị nhiều môn thể thao, đặc biệt là điền kinh và bơi lội. Điều này một phần là nhờ vào hệ thống đào tạo và phát triển tài năng mạnh mẽ của quốc gia này.

Các quốc gia như Mỹ, Nga (hoặc Liên Xô trước đó), Trung Quốc, và các nước châu Âu thường thống trị Thế vận hội về số lượng huy chương, nhờ vào hệ thống huấn luyện thể thao mạnh mẽ và đầu tư lớn.


```python
total_medals = olympics_df_medal['Team'].value_counts()

medals_by_type = olympics_df_medal.groupby(['Team', 'Medal']).size().unstack(fill_value=0)

top_3_countries = total_medals.nlargest(3)
```

Olympic Games cũng phản ánh năng lực thể thao của các quốc gia. Có những quốc gia vượt trội trong nhiều môn thể thao, trong khi những nước khác chỉ nổi bật trong những lĩnh vực nhất định

```python
medals_per_year_country = olympics_df_medal.groupby(['Year', 'Team']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 8))
medals_per_year_country[top_3_countries.index].plot(marker='o', linestyle='-', figsize=(12, 8))
plt.title('Xu hướng huy chương của top 3 quốc gia qua thời gian')
plt.xlabel('Năm')
plt.ylabel('Số huy chương')
plt.legend(title='Quốc Gia (Team)')
plt.grid(True)
plt.show()
```

Liên bang Cộng hòa Xã hội chủ nghĩa Xô viết (USSR) lần đầu tiên tham gia Thế vận hội Olympic vào năm 1952, đội tuyển Liên Xô đứng đầu về tổng số huy chương vàng giành được, đứng thứ hai ba lần và trở thành ứng cử viên lớn nhất cho sự thống trị của Hoa Kỳ tại Thế vận hội Mùa hè. Thành công của Liên Xô có thể là nhờ vào khoản đầu tư lớn của nhà nước vào thể thao để hoàn thành các mục tiêu chính trị của mình trên trường quốc tế.


Năm 1980, Liên Xô đã trở thành quốc gia đăng cai Thế vận hội Mùa hè tại Moscow . Hoa Kỳ và nhiều quốc gia khác đã tẩy chay Thế vận hội này để phản đối cuộc xâm lược Afghanistan của Liên Xô


Mặc dù Liên Xô không còn tồn tại vào ngày 26 tháng 12 năm 1991 nhưng Ủy ban Olympic của Liên Xô vẫn tồn tại chính thức cho đến ngày 12 tháng 3 năm 1992, khi nó giải tán.

```python
top10_athletes=olympics_df_medal['Name'].value_counts().head(10)
top10_athletes
```

```python
top_athlete = top10_athletes.idxmax()
top_medals = top10_athletes.max()
top_athlete
```

Michael Fred Phelps II OLY (sinh ngày 30 tháng 6 năm 1985) là một cựu vận động viên bơi lội chuyên nghiệp người Mỹ. Anh là vận động viên Olympic thành công nhất và giành được nhiều huy chương nhất mọi thời đại với tổng cộng 28 huy chương.

```python
sports_counts = data['Sport'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=sports_counts.values, y=sports_counts.index, palette='viridis')
plt.title('10 môn thể thao được chơi nhiều nhất')
plt.xlabel('')
plt.ylabel('sports')
plt.show()
```

Một số môn thể thao và sự kiện đã trở thành nền tảng của Olympic, nhờ các môn thể thao này có thể đưa ra những thông tin quý giá.


```python
sports_per_year = data.groupby('Year')['Sport'].nunique()

plt.figure(figsize=(10, 6))
plt.bar(sports_per_year.index, sports_per_year.values, color='g')
plt.title('Số môn thể thao qua các kỳ Olympic')
plt.xlabel('Năm')
plt.ylabel('Số môn thể thao')
plt.show()
```

Biểu đồ cột thể hiện sự gia tăng số lượng môn thể thao được đưa vào Thế vận hội. Điều này phản ánh sự mở rộng và đa dạng hóa các môn thi đấu.

Ngoài các môn thể thao truyền thống như điền kinh, bơi lội, và bóng đá, Olympic đã giới thiệu nhiều môn thể thao mới như trượt ván, karate, và bóng rổ 3x3. Điều này giúp Olympic ngày càng trở nên phong phú và thu hút đông đảo khán giả.


Các năm 1940 và 1944, Thế vận hội Mùa hè và Thế vận hội Mùa đông không tổ chức do ảnh hưởng của chiến tranh thế giới lần thứ hai.
